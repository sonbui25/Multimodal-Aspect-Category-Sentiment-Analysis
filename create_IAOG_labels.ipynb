```python
#  1. C·∫•u h√¨nh Model v√† Prompt Template cho Gemini 
# (Ph·∫ßn n√†y gi·ªØ nguy√™n nh∆∞ c≈© t·ª´ cell tr∆∞·ªõc)

# C·∫•u h√¨nh model
generation_config = {
  "max_output_tokens": 64,
}
safety_settings = [
    {"category": "HARM_CATEGORY_HARASSMENT", "threshold": "BLOCK_NONE"},
    {"category": "HARM_CATEGORY_HATE_SPEECH", "threshold": "BLOCK_NONE"},
    {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_NONE"},
    {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_NONE"},
]
# S·ª≠a l·∫°i t√™n model cho ƒë√∫ng
model = genai.GenerativeModel(model_name="gemini-2.5-flash-lite",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

# Prompt Template (Gi·ªØ nguy√™n)
PROMPT_TEMPLATE = """B·∫°n l√† m·ªôt chuy√™n gia ng√¥n ng·ªØ. H√£y ph√¢n t√≠ch kƒ© 'C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n)' v√† 'Kh√≠a c·∫°nh' v·ªõi b·ªëi c·∫£nh d·ªØ li·ªáu l√† kh√°ch h√†ng nh·∫≠n x√©t v·ªÅ kh√°ch s·∫°n ·ªü c√°c kh√≠a c·∫°nh kh√°c nhau.

GI·∫¢I TH√çCH C√ÅC KH√çA C·∫†NH 
Room: Kh√≠a c·∫°nh n√†y bao g·ªìm c√°c nh·∫≠n x√©t v·ªÅ ph√≤ng kh√°ch s·∫°n, bao g·ªìm c√°c kh√≠a c·∫°nh nh∆∞ k√≠ch th∆∞·ªõc, thi·∫øt k·∫ø, n·ªôi th·∫•t, ph√≤ng t·∫Øm, c√°ch √¢m.
Location: Kh√≠a c·∫°nh n√†y ƒë·ªÅ c·∫≠p ƒë·∫øn v·ªã tr√≠ c·ªßa kh√°ch s·∫°n, c√°c ƒëi·ªÉm tham quan g·∫ßn ƒë√≥.
Food: Kh√≠a c·∫°nh n√†y bao g·ªìm ph·∫£n h·ªìi v·ªÅ b·ªØa s√°ng, ƒë·ªì ƒÉn, ƒë·ªì u·ªëng, buffet
Facilities: Kh√≠a c·∫°nh n√†y bao g·ªìm c√°c ti·ªán nghi m√† kh√°ch s·∫°n cung c·∫•p ƒë·ªÉ ƒë√°p ·ª©ng nhu c·∫ßu v√† s·ªü th√≠ch c√° nh√¢n c·ªßa kh√°ch, ch·∫≥ng h·∫°n nh∆∞ h·ªì b∆°i, ph√≤ng gym, nh√† h√†ng, internet, spa, c√¢u l·∫°c b·ªô, ti·ªán nghi cho tr·∫ª em.
Service: Kh√≠a c·∫°nh n√†y bao g·ªìm c√°c nh·∫≠n x√©t chung v·ªÅ d·ªãch v·ª•, th√°i ƒë·ªô c·ªßa nh√¢n vi√™n, ch·∫•t l∆∞·ª£ng d·ªãch v·ª•, c√°c v·∫•n ƒë·ªÅ li√™n quan ƒë·∫øn th·ªß t·ª•c nh·∫≠n ph√≤ng v√† tr·∫£ ph√≤ng.
Public Area: Kh√≠a c·∫°nh n√†y ƒë·ªÅ c·∫≠p ƒë·∫øn quang c·∫£nh v√† c√°c kh√¥ng gian chung d√†nh cho kh√°ch, bao g·ªìm s·∫£nh, h√†nh lang, khu v∆∞·ªùn, s√¢n trong.

Nhi·ªám v·ª•:
D·ª±a v√†o 'Kh√≠a c·∫°nh', h√£y quy·∫øt ƒë·ªãnh xem t·ª´ n√†o trong '·ª®ng vi√™n' c√≥ th·ªÉ ƒë∆∞·ª£c d√πng ƒë·ªÉ m√¥ t·∫£ v·ªÅ kh√≠a c·∫°nh ho·∫∑c c·∫£m x√∫c c·ªßa ng∆∞·ªùi n√≥i v·ªÅ kh√≠a c·∫°nh ƒë√≥, d·ª±a v√†o ƒë·∫∑c ƒëi·ªÉm theo m·ª©c ƒë·ªô ∆∞u ti√™n sau:
∆Øu ti√™n 1. X√©t to√†n b·ªô t√≠nh t·ª´, n·∫øu c√°c t√≠nh t·ª´ ƒë√≥ c√≥ th·ªÉ ƒë∆∞·ª£c d√πng ƒë·ªÉ m√¥ t·∫£ 'Kh√≠a c·∫°nh' m·ªôt c√°ch ph√π h·ª£p v·ªÅ m·∫∑t ng·ªØ nghƒ©a th√¨ tr·∫£ v·ªÅ c√°c t√≠nh t·ª´ ƒë√≥. V√≠ d·ª•: Trong danh s√°ch '·ª®ng vi√™n' c√≥ t√≠nh t·ª´ 's·∫°ch' v√† 'Kh√≠a c·∫°nh' l√† 'Room' th√¨ tr·∫£ v·ªÅ 's·∫°ch' v√¨ m·ªôt cƒÉn ph√≤ng c√≥ th·ªÉ ƒë∆∞·ª£c m√¥ t·∫£ l√† s·∫°ch s·∫Ω, h·ª£p l√Ω v·ªÅ m·∫∑t ng·ªØ nghƒ©a. N·∫øu t√≠nh t·ª´ ƒë√≥ kh√¥ng c√≥ t√≠nh ch·∫•t n√†y, chuy·ªÉn sang ∆∞u ti√™n 2.
∆Øu ti√™n 2. X√©t t√≠nh t·ª´, ƒë·ªông t·ª´ ƒë√≥ m√¥ t·∫£ c·∫£m x√∫c/th√°i ƒë·ªô c·ªßa ng∆∞·ªùi n√≥i theo h∆∞·ªõng t√≠ch c·ª±c hay ti√™u c·ª±c. V√≠ d·ª•: Danh s√°ch t·ª´ '·ª®ng vi√™n' l√† ['h√†i l√≤ng', 'th·∫•t v·ªçng', 't·ª©c gi·∫≠n', 'ch√™'] v√† 'Kh√≠a c·∫°nh' l√† 'Public Area' th√¨ tr·∫£ v·ªÅ to√†n b·ªô danh s√°ch c√°c t·ª´ '·ª®ng vi√™n' tr√™n do n√≥ th·ªÉ hi·ªán c·∫£m x√∫c c·ªßa ng∆∞·ªùi n√≥i v·ªÅ 'Kh√≠a c·∫°nh' ƒë∆∞·ª£c ƒë·ªÅ c·∫≠p. 

QUY T·∫ÆC TR·∫¢ V·ªÄ (B·∫ÆT BU·ªòC):
- CH·ªà TR·∫¢ V·ªÄ 1 D√íNG DUY NH·∫§T, V·ªöI 1 LIST PYTHON duy nh·∫•t, v√≠ d·ª•: ['s·∫°ch', 'r·ªông r√£i']
- KH√îNG GI·∫¢I TH√çCH, KH√îNG GHI D√íNG PH·ª§, KH√îNG L·∫∂P
- N·∫øu kh√¥ng c√≥ t·ª´ n√†o h·ª£p l·ªá, CH·ªà TR·∫¢ V·ªÄ: []

M·ªòT S·ªê V√ç D·ª§ M·∫™U

V√≠ d·ª• 1 (L·ªçc b·ªè ng·ªØ c·∫£nh sai):
Kh√≠a c·∫°nh: "Room"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): ['s·∫°ch', 'ngon']
K·∫øt qu·∫£:
['s·∫°ch']

V√≠ d·ª• 2 (Th√°i ƒë·ªô t√≠ch c·ª±c):
Kh√≠a c·∫°nh: "Public Area"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): ['h√†i l√≤ng']
K·∫øt qu·∫£:
['h√†i l√≤ng']

V√≠ d·ª• 3 (Th√°i ƒë·ªô ti√™u c·ª±c):
Kh√≠a c·∫°nh: "Service"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): ['th·∫•t v·ªçng', 't·ªá']
K·∫øt qu·∫£:
['th·∫•t v·ªçng', 't·ªá']

V√≠ d·ª• 4 (L·ªçc t√≠nh t·ª´ ƒë·ªÉ ph√π h·ª£p theo kh√≠a c·∫°nh):
Kh√≠a c·∫°nh: "Food"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): ['chu ƒë√°o', 'tuy·ªát v·ªùi']
K·∫øt qu·∫£:
['tuy·ªát v·ªùi']

V√≠ d·ª• 5 (ƒê·ªëi v·ªõi c√°c tr∆∞·ªùng h·ª£p ph·ªß ƒë·ªãnh, ch·ªâ tr·∫£ v·ªÅ t√≠nh t·ª´ ph√π h·ª£p ƒë·ªÉ m√¥ t·∫£ kh√≠a c·∫°nh)
Kh√≠a c·∫°nh: "Public Area"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): ['s·∫°ch', 'r·ªông r√£i']
K·∫øt qu·∫£:
['s·∫°ch', 'r·ªông r√£i']

V√≠ d·ª• 6 (ƒê·ªông t·ª´ mang √Ω nghƒ©a c·∫£m x√∫c)
Kh√≠a c·∫°nh: "Facilities"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): ['ch√™']
K·∫øt qu·∫£:
['ch√™']
K·∫æT TH√öC V√ç D·ª§, tr·∫£ v·ªÅ k·∫øt qu·∫£ cho ƒë·∫ßu v√†o nh∆∞ sau:
Kh√≠a c·∫°nh: "{aspect}"
C√°c t·ª´ c·∫£m x√∫c (·ª®ng vi√™n): {candidates}
K·∫øt qu·∫£:
"""

# H√†m call_gemini_to_match (Gi·ªØ nguy√™n)a
def call_gemini_to_match(aspect, candidates):
    """
    H√†m g·ªçi API Gemini ƒë·ªÉ l·ªçc c√°c ·ª©ng vi√™n
    """
    if not candidates: # Kh√¥ng g·ªçi API n·∫øu kh√¥ng c√≥ ·ª©ng vi√™n
        return []

    # ƒê·ªãnh d·∫°ng prompt
    prompt = PROMPT_TEMPLATE.format(
        aspect=aspect,
        candidates= candidates # v√≠ d·ª•: "['s·∫°ch', 'ngon']"
    )
    
    try:
        # G·ªçi API
        response = model.generate_content(prompt)
        response_text = response.text.strip()
        # D√πng regex ƒë·ªÉ tr√≠ch xu·∫•t an to√†n c√°c t·ª´ (tr√°nh d√πng eval())
        matched_words = re.findall(r"'(.*?)'", response_text)
        # Ch·ªâ gi·ªØ l·∫°i c√°c t·ª´ m√† LLM tr·∫£ v·ªÅ c√≥ trong danh s√°ch ·ª©ng vi√™n g·ªëc
        valid_words = [word for word in matched_words if word in candidates]
        return valid_words
    
    except Exception as e:
        print(f"L·ªói khi tr√≠ch xu·∫•t t√≠nh t·ª´ cho aspect: '{aspect}': {e}")
        return [] # Tr·∫£ v·ªÅ r·ªóng n·∫øu l·ªói

#  2. H√†m T·∫°o Nh√£n

# H√†m generate_iaog_pairs_with_logging (thay th·∫ø print b·∫±ng tqdm.write)
def generate_iaog_pairs_with_logging(row, lexicon_pos, lexicon_neg):
    """
    T·∫°o m·ªôt LIST c√°c c·∫∑p (sentiment, aspect) v·ªõi logging ph√π h·ª£p cho tqdm.
    """
    # 1. T√¨m Aspect Ng·∫ßm
    all_text_one_img_labels = set()
    for labels in row['text_one_img_label']:
        all_text_one_img_labels.update(labels)

    text_only_aspects = set(label.split('#')[0] for label in row['text_label'])
    implicit_labels = all_text_one_img_labels - set(row['text_label'])
    implicit_aspects_with_polarity = [(label.split('#')[0], label.split('#')[1]) for label in implicit_labels]
    if not implicit_aspects_with_polarity:
        return []

    # 2. L·∫•y ·ª®ng vi√™n T√≠nh t·ª´
    comment = row['comment']
    tagged_words = pos_tag(comment)
    adj_candidates = set(word.lower() for word, tag in tagged_words if tag == 'A')
    if not adj_candidates:
        adj_candidates = set(word.lower() for word, tag in tagged_words if tag == 'V' and (word in lexicon_pos or word in lexicon_neg))
    final_pairs = []
    all_filtered_candidates = []

    # 3. L·∫∑p qua t·ª´ng aspect ng·∫ßm
    for aspect, polarity in implicit_aspects_with_polarity:
        if polarity == "Neutral":
            continue

        if len(adj_candidates) == 1:
            filtered_candidates = list(adj_candidates)
        else:
            target_lexicon = lexicon_pos if polarity == "Positive" else lexicon_neg
            filtered_candidates = list(adj_candidates.intersection(target_lexicon))

        all_filtered_candidates.extend(filtered_candidates)
        if not filtered_candidates:
            continue

        matched_opinions = call_gemini_to_match(aspect, filtered_candidates)
        if matched_opinions:
            for opinion in matched_opinions:
                final_pairs.append(f"{opinion}#{aspect}")

    # Logging b·∫±ng tqdm.write
    tqdm.write(f"Comment: {comment}")
    tqdm.write(f"List of filtered candidates: {list(set(all_filtered_candidates))}")
    tqdm.write(f"Implicit aspects with polarity: {implicit_aspects_with_polarity}")
    tqdm.write(f"Final (opinion, aspect) pairs: {final_pairs}\n")

    time.sleep(4.5)  # Gi·ªØ nguy√™n th·ªùi gian ch·ªù
    return final_pairs
```

```python
print("ƒêang t·∫°o c√°c c·∫∑p IAOG (Hybrid-LLM)...")
# √Åp d·ª•ng h√†m m·ªõi. C·ªôt n√†y s·∫Ω ch·ª©a List (c√°c c·∫∑p) ho·∫∑c None.
train_data['iaog_labels'] = train_data.progress_apply(
    lambda row: generate_iaog_pairs_with_logging(row, lexicon_pos, lexicon_neg), 
    axis=1
)
print("Ho√†n t·∫•t t·∫°o c·∫∑p.")

# Ki·ªÉm tra k·∫øt qu·∫£
print("\n--- Ki·ªÉm tra 5 h√†ng ƒë·∫ßu ra (c·ªôt 'iaog_labels') ---")
display(train_data[['comment', 'iaog_labels']].head())
```
