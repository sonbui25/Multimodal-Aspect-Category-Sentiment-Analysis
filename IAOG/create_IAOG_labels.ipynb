```python
#  1. Cấu hình Model và 
# mpt Template cho Gemini 
# (Phần này giữ nguyên như cũ từ cell trước)

# Cấu hình model
generation_config = {
  "max_output_tokens": 4096,
}
safety_settings = [
    {"category": "HARM_CATEGORY_HARASSMENT", "threshold": "BLOCK_NONE"},
    {"category": "HARM_CATEGORY_HATE_SPEECH", "threshold": "BLOCK_NONE"},
    {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_NONE"},
    {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_NONE"},
]
# Sửa lại tên model cho đúng
model = genai.GenerativeModel(model_name="gemini-2.5-flash-lite",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

# Prompt Template (Giữ nguyên)
PROMPT_TEMPLATE = """Bạn là một chuyên gia ngôn ngữ. Hãy phân tích kĩ 'Các từ cảm xúc (Ứng viên)' và 'Khía cạnh' với bối cảnh dữ liệu là khách hàng nhận xét về khách sạn ở các khía cạnh khác nhau.

GIẢI THÍCH CÁC KHÍA CẠNH 
Room: Khía cạnh này bao gồm các nhận xét về phòng khách sạn, bao gồm các khía cạnh như kích thước, thiết kế, nội thất, phòng tắm, cách âm.
Location: Khía cạnh này đề cập đến vị trí của khách sạn, các điểm tham quan gần đó.
Food: Khía cạnh này bao gồm phản hồi về bữa sáng, đồ ăn, đồ uống, buffet
Facilities: Khía cạnh này bao gồm các tiện nghi mà khách sạn cung cấp để đáp ứng nhu cầu và sở thích cá nhân của khách, chẳng hạn như hồ bơi, phòng gym, nhà hàng, internet, spa, câu lạc bộ, tiện nghi cho trẻ em.
Service: Khía cạnh này bao gồm các nhận xét chung về dịch vụ, thái độ của nhân viên, chất lượng dịch vụ, các vấn đề liên quan đến thủ tục nhận phòng và trả phòng.
Public Area: Khía cạnh này đề cập đến quang cảnh và các không gian chung dành cho khách, bao gồm sảnh, hành lang, khu vườn, sân trong.

Nhiệm vụ:
Dựa vào 'Khía cạnh', hãy quyết định xem từ nào trong 'Ứng viên' có thể được dùng để mô tả về khía cạnh hoặc cảm xúc của người nói về khía cạnh đó, dựa vào đặc điểm theo mức độ ưu tiên sau:
Ưu tiên 1. Xét toàn bộ tính từ, nếu các tính từ đó có thể được dùng để mô tả 'Khía cạnh' một cách phù hợp về mặt ngữ nghĩa thì trả về các tính từ đó. Ví dụ: Trong danh sách 'Ứng viên' có tính từ 'sạch' và 'Khía cạnh' là 'Room' thì trả về 'sạch' vì một căn phòng có thể được mô tả là sạch sẽ, hợp lý về mặt ngữ nghĩa. Nếu tính từ đó không có tính chất này, chuyển sang ưu tiên 2.
Ưu tiên 2. Xét tính từ, động từ đó mô tả cảm xúc/thái độ của người nói theo hướng tích cực hay tiêu cực. Ví dụ: Danh sách từ 'Ứng viên' là ['hài lòng', 'thất vọng', 'tức giận', 'chê'] và 'Khía cạnh' là 'Public Area' thì trả về toàn bộ danh sách các từ 'Ứng viên' trên do nó thể hiện cảm xúc của người nói về 'Khía cạnh' được đề cập. 

QUY TẮC TRẢ VỀ (BẮT BUỘC):
- CHỈ TRẢ VỀ 1 DÒNG DUY NHẤT, VỚI 1 LIST PYTHON duy nhất, ví dụ: ['sạch', 'rộng rãi']
- KHÔNG GIẢI THÍCH, KHÔNG GHI DÒNG PHỤ, KHÔNG LẶP
- Nếu không có từ nào hợp lệ, CHỈ TRẢ VỀ: []

MỘT SỐ VÍ DỤ MẪU

Ví dụ 1 (Lọc bỏ ngữ cảnh sai):
Khía cạnh: "Room"
Các từ cảm xúc (Ứng viên): ['sạch', 'ngon']
Kết quả:
['sạch']

Ví dụ 2 (Thái độ tích cực):
Khía cạnh: "Public Area"
Các từ cảm xúc (Ứng viên): ['hài lòng']
Kết quả:
['hài lòng']

Ví dụ 3 (Thái độ tiêu cực):
Khía cạnh: "Service"
Các từ cảm xúc (Ứng viên): ['thất vọng', 'tệ']
Kết quả:
['thất vọng', 'tệ']

Ví dụ 4 (Lọc tính từ để phù hợp theo khía cạnh):
Khía cạnh: "Food"
Các từ cảm xúc (Ứng viên): ['chu đáo', 'tuyệt vời']
Kết quả:
['tuyệt vời']

Ví dụ 5 (Đối với các trường hợp phủ định, chỉ trả về tính từ phù hợp để mô tả khía cạnh)
Khía cạnh: "Public Area"
Các từ cảm xúc (Ứng viên): ['sạch', 'rộng rãi']
Kết quả:
['sạch', 'rộng rãi']

Ví dụ 6 (Động từ mang ý nghĩa cảm xúc)
Khía cạnh: "Facilities"
Các từ cảm xúc (Ứng viên): ['chê']
Kết quả:
['chê']
KẾT THÚC VÍ DỤ, trả về kết quả cho đầu vào như sau:
Khía cạnh: "{aspect}"
Các từ cảm xúc (Ứng viên): {candidates}
Kết quả:
"""

# Hàm call_gemini_to_match (Giữ nguyên)a
def call_gemini_to_match(aspect, candidates):
    """
    Hàm gọi API Gemini để lọc các ứng viên
    """
    if not candidates: # Không gọi API nếu không có ứng viên
        return []

    # Định dạng prompt
    prompt = PROMPT_TEMPLATE.format(
        aspect=aspect,
        candidates= candidates # ví dụ: "['sạch', 'ngon']"
    )
    
    try:
        # Gọi API
        response = model.generate_content(prompt)
        response_text = response.text.strip()
        # Dùng regex để trích xuất an toàn các từ (tránh dùng eval())
        matched_words = re.findall(r"'(.*?)'", response_text)
        # Chỉ giữ lại các từ mà LLM trả về có trong danh sách ứng viên gốc
        valid_words = [word for word in matched_words if word in candidates]
        return valid_words
    
    except Exception as e:
        print(f"Lỗi khi trích xuất tính từ cho aspect: '{aspect}': {e}")
        return [] # Trả về rỗng nếu lỗi

#  2. Hàm Tạo Nhãn

# Hàm generate_iaog_pairs_with_logging (thay thế print bằng tqdm.write)
def generate_iaog_pairs_with_logging(row, lexicon_pos, lexicon_neg):
    """
    Tạo một LIST các cặp (sentiment, aspect) với logging phù hợp cho tqdm.
    """
    # 1. Tìm Aspect Ngầm
    all_text_one_img_labels = set()
    for labels in row['text_one_img_label']:
        all_text_one_img_labels.update(labels)

    text_only_aspects = set(label.split('#')[0] for label in row['text_label'])
    implicit_labels = all_text_one_img_labels - set(row['text_label'])
    implicit_aspects_with_polarity = [(label.split('#')[0], label.split('#')[1]) for label in implicit_labels]
    if not implicit_aspects_with_polarity:
        return []

    # 2. Lấy Ứng viên Tính từ
    comment = row['comment']
    tagged_words = pos_tag(comment)
    adj_candidates = set(word.lower() for word, tag in tagged_words if tag == 'A')
    if not adj_candidates:
        adj_candidates = set(word.lower() for word, tag in tagged_words if tag == 'V' and (word in lexicon_pos or word in lexicon_neg))
    final_pairs = []
    all_filtered_candidates = []

    # 3. Lặp qua từng aspect ngầm
    for aspect, polarity in implicit_aspects_with_polarity:
        if polarity == "Neutral":
            continue

        if len(adj_candidates) == 1:
            filtered_candidates = list(adj_candidates)
        else:
            target_lexicon = lexicon_pos if polarity == "Positive" else lexicon_neg
            filtered_candidates = list(adj_candidates.intersection(target_lexicon))

        all_filtered_candidates.extend(filtered_candidates)
        if not filtered_candidates:
            continue

        matched_opinions = call_gemini_to_match(aspect, filtered_candidates)
        if matched_opinions:
            for opinion in matched_opinions:
                final_pairs.append(f"{opinion}#{aspect}")

    # Logging bằng tqdm.write
    # tqdm.write(f"Comment: {comment}")
    # tqdm.write(f"List of filtered candidates: {list(set(all_filtered_candidates))}")
    # tqdm.write(f"Implicit aspects with polarity: {implicit_aspects_with_polarity}")
    # tqdm.write(f"Final (opinion, aspect) pairs: {final_pairs}\n")

    # time.sleep(1)  # Giữ nguyên thời gian chờ
    return final_pairs
```

```python
print("Đang tạo các cặp IAOG (Hybrid-LLM)...")
# Áp dụng hàm mới. Cột này sẽ chứa List (các cặp) hoặc None.
df['iaog_labels'] = df.progress_apply(
    lambda row: generate_iaog_pairs_with_logging(row, lexicon_pos, lexicon_neg), 
    axis=1
)
print("Hoàn tất tạo cặp.")

# Kiểm tra kết quả
print("\n--- Kiểm tra 5 hàng đầu ra (cột 'iaog_labels') ---")
display(df[['comment', 'iaog_labels']].head())
```
